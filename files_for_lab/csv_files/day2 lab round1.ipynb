```python
import pandas as pd
```

```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')

cols1=[]
for c in file1.columns:
    cols1.append(c.lower())
file1.columns=cols1

cols2=[]
for c in file2.columns:
    cols2.append(c.lower())
file2.columns=cols2

cols3=[]
for c in file3.columns:
    cols3.append(c.lower())
file3.columns=cols3

file1 = file1[['customer','st','gender','education','customer lifetime value','monthly premium auto','vehicle class','policy type','number of open complaints','income','total claim amount']]
file1.head()
```

```python
file2 = file2[['customer','st','gender','education','customer lifetime value','monthly premium auto','vehicle class','policy type','number of open complaints','income','total claim amount']]
file2.head()
```

```python
file3 = file3.rename(columns={'state':'st'})
file3 = file3[['customer','st','gender','education','customer lifetime value','monthly premium auto','vehicle class','policy type','number of open complaints','income','total claim amount']]
file3.head()
```

```python
buyers = pd.concat([file1,file2,file3], axis=0)
buyers.shape
```

```python
buyers = buyers.drop_duplicates()
#buyers = buyers.drop(['education','number of open frames'], axis=1)

buyers
```

```python
#buyers['customer lifetime value'] = pd.to_numeric(buyers['customer lifetime value'], errors='coerce')

clv_corrected = list()
for value in buyers['customer lifetime value']:
      if '%' in str(value):
             value = value[:-1]
             clv_corrected.append(val)
      else:
             clv_corrected.append(val)

buyers['customer lifetime value'] = clv_corrected



buyers
```
